```python
from scipy.io import arff
import pandas as pd
import numpy as np
```

```python
data = arff.loadarff('4year.arff')
df = pd.DataFrame(data[0])
```

```python
df['bankruptcy'] = (df['class']==b'1')
del df['class']
df
```

### Create a new dataframe with only 4 feataures (and and Bankruptcy). Properly rename the columns to X1, X2, X7, and X10

```python
df_new = df[['Attr1', 'Attr2', 'Attr7', 'Attr10', 'bankruptcy']]
df_new.columns = ['X1', 'X2', 'X7', 'X10', 'bankruptcy']
df_new
```

### Fill-in the missing values (nan) with the column means. (Use pd.fillna() or See Ch 4 of PML)

```python
# there are nans in dataframe
df_new.isnull().sum()
```

```python
# impute missing values via the column mean
df_new.fillna(df_new.mean(), inplace=True)
```

```python
df_new.isnull().sum()
```

### Find the mean and std of the 4 features among all, bankrupt and still-operating companies (3 groups).

```python
print(df_new.mean())
print(df_new.std())
```

```python
print(df_new.groupby('bankruptcy').mean())
print(df_new.groupby('bankruptcy').std())
```

### How many companies satisfy the condition, X1 < mean(X1) - stdev(X1) AND X10 < mean(X10) - std(X10)?

```python
a = df_new['X1'].mean() - df_new['X1'].std()
b = df_new['X10'].mean() - df_new['X10'].std()
df_satisfy = df_new[(df_new['X1'] < a) & (df_new['X10'] < b)]
df_satisfy
```

```python
print("There are" + str(len(df_satisfy)) + "companies satisfy the condition")
```

### What is the ratio of the bankrupted companies among the sub-groups above?

```python
len(df_satisfy[df_satisfy['bankruptcy']==0])/15
```
